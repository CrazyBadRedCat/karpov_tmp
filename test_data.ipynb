```python
post_features = post_data.copy()
post_features['text_len'] = post_features['text'].str.len()
post_features['word_count'] = post_features['text'].str.split().str.len()
post_features = post_features[['post_id', 'topic', 'text_len', 'word_count', 'text']]
```

```python
def compute_user_agg_advanced(history_df, post_features):
    """
    history_df: DataFrame с колонками user_id, post_viewed (list), post_liked (list)
    Возвращает DataFrame с агрегатами: 
        - общие и уникальные просмотры/лайки,
        - средние длины текста,
        - количество просмотров/лайков по каждой теме,
        - доля лайков по темам,
        - самый популярный топик.
    """
    # Разворачиваем списки
    views = history_df[['user_id', 'post_viewed']].explode('post_viewed')
    views = views.rename(columns={'post_viewed': 'post_id'}).dropna(subset=['post_id'])
    views = views.merge(post_features, on='post_id', how='left')

    likes = history_df[['user_id', 'post_liked']].explode('post_liked')
    likes = likes.rename(columns={'post_liked': 'post_id'}).dropna(subset=['post_id'])
    likes = likes.merge(post_features, on='post_id', how='left')

    # Базовые агрегаты по просмотрам
    views_agg = views.groupby('user_id').agg(
        total_views=('post_id', 'count'),
        unique_views=('post_id', 'nunique'),
        avg_text_len_viewed=('text_len', 'mean'),
        avg_word_count_viewed=('word_count', 'mean')
    )

    # Базовые агрегаты по лайкам
    likes_agg = likes.groupby('user_id').agg(
        total_likes=('post_id', 'count'),
        unique_likes=('post_id', 'nunique'),
        avg_text_len_liked=('text_len', 'mean'),
        avg_word_count_liked=('word_count', 'mean')
    )

    # Количество просмотров по темам
    topic_view_counts = views.groupby(['user_id', 'topic']).size().unstack(fill_value=0)
    topic_view_counts.columns = [f'view_topic_{col}' for col in topic_view_counts.columns]

    # Количество лайков по темам
    topic_like_counts = likes.groupby(['user_id', 'topic']).size().unstack(fill_value=0)
    topic_like_counts.columns = [f'like_topic_{col}' for col in topic_like_counts.columns]


    # Самый популярный топик (по просмотрам)
    most_popular = topic_view_counts.idxmax(axis=1).rename('most_common_topic').str.replace("view_topic_", "")
    most_popular = most_popular.replace(0, 'unknown')  # если нет просмотров

    # Объединяем все в один DataFrame
    user_agg = pd.concat([
        views_agg, likes_agg,
        topic_view_counts, topic_like_counts,
        most_popular
    ], axis=1).reset_index()

    # Заполняем пропуски (для пользователей без активности в каком-либо разрезе)
    fill_zero_cols = ['total_views', 'unique_views', 'total_likes', 'unique_likes'] + \
                     list(topic_view_counts.columns) + list(topic_like_counts.columns)
    for col in fill_zero_cols:
        if col in user_agg.columns:
            user_agg[col] = user_agg[col].fillna(0).astype(int)  # для счётчиков int
    # Средние длины оставляем NaN – заполним позже медианами по выборке
    return user_agg
```

```python
def create_samples_advanced(data_df, post_features, user_agg, user_data, random_state=42):
    """
    Формирует сбалансированный датасет пар (user_id, post_id) с target = 0/1.
    Добавляет:
        - фичи поста (topic, text_len, word_count),
        - демографические фичи пользователя,
        - агрегаты из истории (user_agg),
        - глобальные статистики по постам,
        - признаки взаимодействия (совпадение темы, разница длин).
    """
    # Позитивы (уникальные пары пользователь-пост)
    pos = data_df[['user_id', 'post_liked']].explode('post_liked')
    pos = pos.rename(columns={'post_liked': 'post_id'}).dropna(subset=['post_id'])
    pos = pos.drop_duplicates(subset=['user_id', 'post_id'])
    pos['target'] = 1

    # Все просмотры (кандидаты в негативы)
    views = data_df[['user_id', 'post_viewed']].explode('post_viewed')
    views = views.rename(columns={'post_viewed': 'post_id'}).dropna(subset=['post_id'])
    views = views.drop_duplicates(subset=['user_id', 'post_id'])

    # Исключаем позитивы
    pos_set = set(zip(pos['user_id'], pos['post_id']))
    neg_mask = ~pd.Series(zip(views['user_id'], views['post_id'])).isin(pos_set)
    neg = views[neg_mask].copy()
    neg['target'] = 0

    # Балансировка: столько же негативов, сколько позитивов
    n_pos = len(pos)
    if n_pos == 0:
        return pd.DataFrame()
    n_neg = min(len(neg), n_pos)
    neg_sampled = neg.sample(n=n_neg, random_state=random_state)

    samples = pd.concat([pos, neg_sampled], ignore_index=True)

    # Добавляем фичи поста
    samples = samples.merge(post_features, on='post_id', how='left')

    # Добавляем демографию пользователя
    samples = samples.merge(user_data, on='user_id', how='left')

    # Добавляем агрегаты из истории
    samples = samples.merge(user_agg, on='user_id', how='left')

    # Признаки взаимодействия
    samples['topic_match'] = (samples['topic'] == samples['most_common_topic']).astype(int)
    samples['len_diff_from_avg_view'] = samples['text_len'] - samples['avg_text_len_viewed']
    samples['len_ratio_from_avg_view'] = samples['text_len'] / (samples['avg_text_len_viewed'] + 1)

    # Для пользователей без истории (avg_text_len_viewed = NaN) заполним медианой позже
    return samples
```

```python
train_user_agg = compute_user_agg_advanced(data_1, post_features)

history_12 = pd.concat([data_1, data_2], ignore_index=True)
history_12 = history_12.groupby('user_id').agg(
    post_viewed=('post_viewed', lambda x: sum(x, [])),
    post_liked=('post_liked', lambda x: sum(x, []))
).reset_index()

val_user_agg = compute_user_agg_advanced(history_12, post_features)
```

```python
train_user_agg.to_csv('train_user_agg.csv', index=False)
val_user_agg.to_csv('val_user_agg.csv', index=False)
print("Датасеты сохранены: train_user_agg.csv, val_user_agg.csv")
```

```python
print("Формируем train датасет...")
train_samples = create_samples_advanced(
    data_2, post_features, train_user_agg, user_data
)
print(f"Train samples shape: {train_samples.shape}")
```

```python
print("Формируем validation датасет...")
val_samples = create_samples_advanced(
    data_3, post_features, val_user_agg, user_data
)
print(f"Validation samples shape: {val_samples.shape}")
```

```python
num_cols = [
    'avg_text_len_viewed', 'avg_word_count_viewed',
    'avg_text_len_liked', 'avg_word_count_liked',
    'len_diff_from_avg_view', 'len_ratio_from_avg_view',
] + [col for col in train_samples.columns if col.startswith(('view_topic_', 'like_topic_'))]
```

```python
for col in num_cols:
    if col in train_samples.columns:
        median_val = train_samples[col].median()
        train_samples[col] = train_samples[col].fillna(median_val)
        if col in val_samples.columns:
            val_samples[col] = val_samples[col].fillna(median_val)
```

```python
cat_cols = ['most_common_topic']  # topic уже есть и не содержит NaN после merge
for col in cat_cols:
    if col in train_samples.columns:
        mode_val = train_samples[col].mode()[0] if not train_samples[col].mode().empty else 'unknown'
        train_samples[col] = train_samples[col].fillna(mode_val)
        if col in val_samples.columns:
            val_samples[col] = val_samples[col].fillna(mode_val)
```

```python
train_samples.to_csv('train_dataset_advanced.csv', index=False)
val_samples.to_csv('val_dataset_advanced.csv', index=False)
print("Датасеты сохранены: train_dataset_advanced.csv, val_dataset_advanced.csv")
```
